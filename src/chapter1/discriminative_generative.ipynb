```python
import numpy as np
from matplotlib import pyplot as plt

# Three data points
x = np.array([1, 2, 3])
y = np.array([2, 4, 5])

plt.plot(x, y, 'o')
plt.show()
```

```python
# --- Discriminative regression (y on x) ---
A = np.vstack([x, np.ones(len(x))]).T
w, b = np.linalg.lstsq(A, y, rcond=None)[0]
A, w, b
```

```python
y_hat = x*w+b
plt.plot(x, y_hat)
plt.plot(x, y, 'o')
plt.title("Predict y from x")
plt.show()
```

```python
# --- Generative regression (x on y) ---
B = np.vstack([y, np.ones(len(y))]).T
phi1, phi0 = np.linalg.lstsq(B, x, rcond=None)[0]
B, phi1, phi0
```

```python
x_hat = y*phi1+phi0
plt.plot(x_hat, y)
plt.plot(x, y, 'o')
plt.title("Predict x from y")
plt.show()
```

```python
y_hat_gen = (x - phi0)/phi1
plt.plot(x, y_hat_gen)
plt.plot(x, y, 'o')
plt.title("Predict y from x (Generative model)")
plt.show()
```
